```python
# Configuraçõe# Bibliotecas padrão
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as ticker
import requests

import warnings

# Desabilitar todos os warnings
warnings.filterwarnings("ignore")

Processamento = pd.read_csv('Dados/Dados_Embrapa/Processamento/Tbl_ProcessaViniferas.csv', sep=';', encoding='utf-8')
Processamento.head()
```

```python
# Manter primeiras 3 colunas e as últimas 15 colunas


Processamento_tratado = Processamento.iloc[:,-15:].replace(['nd', '*'], np.nan)
Processamento_tratado = pd.concat([Processamento.iloc[:, :3], Processamento_tratado.iloc[:, -15:]], axis=1)
# Adiciona a coluna 'categoria' com valores vazios inicialmente
Processamento_tratado['categoria'] = None
Processamento_tratado['2023'] = Processamento_tratado['2023'].astype(str).str.replace(',', '.').replace(['nd','*'],'0')
Processamento_tratado['2023'] = pd.to_numeric(Processamento_tratado['2023'], errors='coerce')
# Preenche os valores com base na coluna 'control'
Processamento_tratado.loc[Processamento_tratado['control'].str.startswith('ti_'), 'categoria'] = 'TINTAS'
Processamento_tratado.loc[Processamento_tratado['control'].str.startswith('br_'), 'categoria'] = 'BRANCAS E ROSADAS'

colunas_para_manter = ['categoria','cultivar'] + Processamento_tratado.dropna(axis='columns').columns.tolist()
Processamento_final = Processamento_tratado[colunas_para_manter]
Processamento_final = Processamento_tratado.dropna(subset=['categoria'])
Processamento_final = Processamento_final.drop(columns=['control'])
Processamento_final = Processamento_final.loc[:, ~Processamento_final.columns.duplicated()]
Processamento_final.head(40)

```

```python
Processamento_fim = pd.melt(Processamento_final, id_vars=['categoria', 'cultivar', 'id'], value_vars=Processamento_final.iloc[:, -15:], var_name="Ano", value_name="Total_Qtd_KG")

# Filtrar apenas linhas onde Ano é '2023'
Processamento_fim_2023 = Processamento_fim[Processamento_fim['Ano'] == '2023']
Processamento_fim_2023
```
